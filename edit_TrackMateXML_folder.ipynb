```python
def changeTrackMateSettingsAndSave(xmlfile, output_xmlfile, *args):
    try:
        tree = et.parse(xmlfile)
    except OSError:
        print('Failed to read XML file {}.'.format(xmlfile) )
    
    root =  tree.getroot()
    settings_el = root.find("Settings")

    for change_dict in args:
        settings_el[change_dict["Settings_Index"]].set(change_dict["Key_Attribute"], change_dict["Value_Replacement"])
    #settings_el[0].set("folder", r"C:\Users\wgiang\Dsg1-tracking\02c_cropped-ROIS_renamed_DC_3s")
    #settings_el[2].set("CLASSIFIER_FILEPATH", r"C:\Users\wgiang\Dsg1-tracking\ilastik\2024-02-21_pixel-classification-Dsg1-puncta.ilp")

    tree.write(output_xmlfile, encoding = "unicode")

output_dir_exists = os.path.exists(output_xml_dir)

if not output_dir_exists:
    os.makedirs(output_xml_dir)

input_xmls = os.listdir(input_xml_dir)

for f in input_xmls:
    print(f)
    f_path = os.path.join(input_xml_dir, f)

    out_xml_path = os.path.join(output_xml_dir, f)
    changeTrackMateSettingsAndSave(f_path, out_xml_path, *all_changes)
```
